## 1. More plotting examples

## Example: Trefoil knot in 2d

the trefoil knot in 2D given by the parametirc equations:
$$
x(t)= \sin(t) + 2\sin(2t) \ \ , \ \  y(t) = \cos(t) - 2\cos(2t)
$$

```python
t = np.linspace(0,2*np.pi,1000)
x = np.sin(t)+2*np.sin(2*t)
y = np.cos(t)-2*np.cos(2*t)
```

```python
plt.plot(x,y)
plt.axis('equal')
plt.xlim([-4,4])
plt.ylim([-4,3])
plt.title('Trefoil Knot in 2D')
```

## Example: Plot several curves together

Plot the curves $y=e^{-x^2} \cos(k x)$ for $k=1,2,5,10$ for xE[-5,5]

```python
x = np.linspace(-5,5,1000)
y1 = np.exp(-x**2) * np.cos(x)
y2 = np.exp(-x**2) * np.cos(2*x)
y3 = np.exp(-x**2) * np.cos(5*x)
y4 = np.exp(-x**2) * np.cos(10*x)
plt.plot(x,y1,x,y2,x,y3,x,y4);
```

Same plot but with a for loop:

```python
x = np.linspace(-2,2,1000)
ks = np.random.rand(10)*10
for k in ks:
    y = np.exp(-x**2) * np.cos(k*x)
    plt.plot(x,y);
```

## 2. Array methods

NumPy arrays have many methods to do transformations and calculations. For example, there is the sum method with simply `sums` the entriees in the array.We access methods using the dot notation following the array variable name.

```python
arr = np.array([1,2,3,4,5])
```

```python
arr.sum()
```

There are equivalent ways of doing the same thing:

```python
np.sum(arr)
```

There is also the prod method with takes the product of the entries:

```python
arr.prod()
```

The method `cumsum` computes the cumulative sum of the entries:

```python
arr.cumsum()
```

For example we can plot a random walk:

```python
x = np.random.rand(1000,10).cumsum(axis=0)
plt.plot(x);
```

We can access the shape of an array using `.shape`

```python
M = np.random.randint(0,2,[4,5])
print(M)
M.shape
```

Notice that we accessed the shape without using paraentheses. This difference between array methods (functions to do with calculations with the array) and array attributes (values associated with the array- like its shape.).

```python
M.sum()
```

```python
M.shape
```

There is also the `transpose` method:

```python
A = np.array([[1,2],[3,4]])
A
```

```python
A.transpose()
```

Also there's the mean:

```python
A.mean()
```

```python
A.mean(axis=0) # axis 0 is across the rows
```

```python
A.mean(axis=1) # axis 1 is down the rows
```

## 3. Array slicing and indexing

We access the entries in an array using the square brackets:

```python
A = np.random.randint(0,10,[3,5])
A
```

```python
A[2,4] # Row index 2 and column index 4
```

```python
A[1,3]
```

To acces a whole row or column we use the colon `:`

```python
A[0,:]
```

```python
A[:,3]
```
