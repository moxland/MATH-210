## 1. Exploring `pyplot` commands

We use matplotlib to generate plots and we use the `pyplot` library of commands to customize our figures. There are many of these [commands](http://matplotlib.org/api/pyplot_summary.html) such as:

* `xlim` and `ylim`
* `grid`
* `title`
* `xlabel` and `ylabel`

Here are some examples:

Let's plot $y=
\sin(x) + \sin(4x)$ for x in range [0,2pi]

```python
x = np.linspace(0,2*np.pi,1000)
y = np.sin(x) + np.sin(4*x)
plt.plot(x,y,'r--',linewidth=2)
plt.xlim(0,2*np.pi), plt.ylim(-3,3), plt.grid('on',linestyle='-.');
plt.title('Sum of sine waves'), plt.xlabel('Time (s)'), plt.ylabel('Amplitude (mm)');
```

There's also a special commands called `show`. We use this if we want to render the current figure. Jupyter notebook calls this for the current figure when it reaches the end of a cell. So we'll use this when we want to render several figures from within the same cell.

```python
x = np.linspace(0,2*np.pi,1000)
y1 = np.sin(x) + np.sin(4*x)
plt.plot(x,y1,'b')
plt.show()

y2 = np.sin(x) + np.sin(4*x) + np.sin(4*x)
plt.plot(x,y2,'r')
plt.show()
```

We can also generate a legend for a figure with several lines. Let's plot $y = \sin(x)$ and $y = \cos(x)$ with a legend:

```python
t = np.linspace(0,4*np.pi,1000)
y1 = np.sin(t)
y2 = np.cos(t)
plt.plot(t,y1,t,y2), plt.xlim(0,4*np.pi), plt.ylim(-2,2);
plt.legend(['y=sin(t)','y=cos(t)'])
```

## 2. Subplots

The command `subplot` takes at least 3 inputs n,m and i and creates a figure with a n by m grid of subplots and then sets the i-th subplots (counting across the rows) as the current plot (ie. current axes option).


```python
for k in range(1,5):
    plt.subplot(2,2,k)
    x = np.linspace(-5,5,1000)
    y = x**(k-1) * np.arctan(x) / (1+x**2)
    plt.plot(x,y),plt.grid('on'), plt.xlim(-5,5), plt.title('k= ' + str(k)), plt.tight_layout()
```

## 3. Different kinds of plots

So far we've just been creating line plots. Let's create some other kinds:

### Scatter Plots

```python
N = 10000
x = np.random.randn(N)
y = np.random.rand(N)
plt.scatter(x,y, lw=0,alpha=0.2,c='purple');
```
