## 3D Plotting

### Example: Saddle

Let's plot $z = x^2 - y^2$ for x and y in [-2,2]

```python
plt.figure()
ax = plt.axes(projection='3d')

x = np.linspace(-2,2,50)
y = np.linspace(-2,2,50)
X,Y = np.meshgrid(x,y)
Z = X**2 - Y**2
ax.contour(X,Y,Z)
```

```python
plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X,Y,Z)
```

```python
plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z)
```

### Example: Trefoil

Let's plot the trefoil knot in 3D:

$$
x = \sin(t) +2\sin(2t) \ \ , \ \ y = \cos(t) - 2\cos(2t) \ \ , \ \ z = -\sin(3t)
$$

```python
plt.figure()
ax = plt.axes(projection='3d')

t = np.linspace(0,2*np.pi,100)

x = np.sin(t) + 2*np.sin(2*t)
y = np.cos(t) - 2*np.cos(2*t)
z = -np.sin(3*t)

ax.view_init(90,0) 
ax.plot(x,y,z);
```

## Exercise

Let's try and plot a 3D torus:

```python
plt.figure()
ax = plt.axes(projection='3d')

R = 3
r = 1
theta = np.linspace(0,2*np.pi,50)
phi = np.linspace(0,2*np.pi,50)

PHI, THETA = np.meshgrid(phi,theta)

x = (R+r*np.cos(THETA))*np.cos(PHI)
y = (R+r*np.cos(THETA))*np.sin(PHI)
z = r*np.sin(THETA)

ax.view_init(60,30)
ax.plot_wireframe(x,y,z)
```

## 2. Miscellaneous advanced topics

### 1. Keyword arguments

We can define funcitons with keyword arguments such as :

```python
def fun(a,b,add=True):
    if add == True:
        return a + b
    else:
        return a * b
```

```python
fun(1,2)
```

In this example `add` imput parameter is called a keyword argument and it has a default value `True`.

The first 2 imput parameters  `a` and `b` are called positional arguments. This means that the order in which values are input into `fun` determine which is `a` and `b`. In other words, in the example above, when we call `fun(1,2)` the function `fun` knows that a=1 and b=2 since that's the order they came.

```python
fun(3,5,add=False)
```

1. $z = (x^2 - y^2)^2$

```python
plt.figure()
ax = plt.axes(projection='3d')

x = np.linspace(-1,1,100)
y = np.linspace(-1,1,100)

X,Y = np.meshgrid(x,y)

Z = (X**2-Y**2)**2

ax.plot_surface(X,Y,Z);
```

2: $z = \sin(|x| + |y|)$

```python
plt.figure()
ax = plt.axes(projection='3d')

x = np.linspace(-4*np.pi,4*np.pi,100)
y = np.linspace(-4*np.pi,4*np.pi,100)

X,Y = np.meshgrid(x,y)

Z = np.sin(np.abs(X)+np.abs(Y))

ax.plot_wireframe(X,Y,Z, alpha=0.5);
```

3: $z = \sin(x) - \sin(y)$

```python
plt.figure()
ax = plt.axes(projection='3d')

x = np.linspace(-4*np.pi,4*np.pi,100)
y = np.linspace(-4*np.pi,4*np.pi,100)

X,Y = np.meshgrid(x,y)

Z = np.sin(X)-np.sin(Y)

ax.plot_surface(X,Y,Z);
```
