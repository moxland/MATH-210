## 1. Matplotlib figure objects and axes objects

When we make a simple plot, there are a few things in the background that we haven't seen yet.
For example, let's plot $y=x^2$


```python
x = np.linspace(-2,2,50)
y = x**2
plt.plot(x,y);
```

When we call the `plot` command, matplotlib creates a new figure object (think of it as a window containing plots) and an axes object (think of it as the axes in which we plot stuff), and then it plots our data to the given axes contained in the given figure.

```python
fig1 = plt.figure() # create a new figure
```

```python
ax1 = fig1.add_subplot(111)
type(ax1)
```

## 2. Plotting in 3D with mplot3d

Let's plot the surface $z=1-x^2-y^2$

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
plt.figure() # creates a new figure
ax = plt.axes(projection='3d') # creates a new axes 3d object in the current figure

x = np.linspace(-1,1,100)
y = np.linspace(-1,1,100)
x, y = np.meshgrid(x,y)
z = 1 - x**2 - y**2
ax.plot_surface(x,y,z);
```

Let's plot $z= \cos( \sqrt{x^2+y^2} )$

```python
plt.figure() # creates a new figure
ax = plt.axes(projection='3d') # creates a new axes 3d object in the current figure

x = np.linspace(-8*np.pi,8*np.pi,500)
y = np.linspace(-8*np.pi,8*np.pi,500)
x, y = np.meshgrid(x,y)
z = np.cos( np.sqrt(x**2 + y**2) )
ax.plot_surface(x,y,z,cmap=plt.cm.winter), ax.set_zlim(bottom=-5,top=5);
```

### What is meshgrid?

Like many 3D plotting libraries (ie. matlab), the 3D plotting functions require 2D matrices of x,y and z values. Meshgrid takes x and y arrays and returns two 2D matrices x and y with x has the values across its rows and y has the y values down its columns.

```python
x = [1,2,3]
y = [8,9,10,11]
np.meshgrid(x,y)
```

The output of meshgrid is a list with 2 entries. We can unpack those entries into variables x and y.

```python
x, y = np.meshgrid(x,y)
```

### Parametric Plot- Sphere

Twp parameters $\theta$ and $\phi$ inspherical coordinates

```python
plt.figure() # creates a new figure
ax = plt.axes(projection='3d') # creates a new axes 3d object in the current figure

phi = np.linspace(0,np.pi,50)
theta = np.linspace(0,2*np.pi,50)

PHI, THETA = np.meshgrid(phi,theta)
x = np.sin(PHI) * np.cos(THETA)
y = np.sin(PHI) * np.sin(THETA)
z = np.cos(PHI)
ax.plot_wireframe(x,y,z,cmap=plt.cm.jet)
```

## Exercises

1. $z=\frac{1}{1+x^2+2y^2}$

```python
plt.figure()
ax = plt.axes(projection='3d')

x=np.linspace(-2,2,100)
y=np.linspace(-2,2,100)

X,Y = np.meshgrid(x,y)

Z = 1/(1 + X**2 + 2*Y**2)

ax.plot_wireframe(X,Y,Z);
```

2: $z=\sqrt{1+x^2+y^2}$

```python
plt.figure()
ax = plt.axes(projection='3d')

x=np.linspace(-2,2,100)
y=np.linspace(-2,2,100)

X,Y = np.meshgrid(x,y)

Z = (1+X**2+Y**2)**0.5

ax.plot_surface(X,Y,Z);
```

3: $z=x^2+y^2$

```python
plt.figure()
ax = plt.axes(projection='3d')

x=np.linspace(-2,2,100)
y=np.linspace(-2,2,100)

X,Y = np.meshgrid(x,y)

Z = X**2-Y**2

ax.plot_surface(X,Y,Z);
```

4: $z = e^{-x^2-y^2} \cos(5(x^2 + y^2))$

```python
plt.figure()
ax = plt.axes(projection='3d')

x=np.linspace(-2,2,100)
y=np.linspace(-2,2,100)

X,Y = np.meshgrid(x,y)

Z = np.exp(-X**2-Y**2)*np.cos(5*(X**2+Y**2))

ax.plot_surface(X,Y,Z);
```
