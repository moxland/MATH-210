```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.linalg as la
import scipy.integrate as spi
```

## Question 1

```python
def exp_integral(a,b,c,n):
    if a<0 or b<0 or c<0 or n<0:
        print("Error: Negative Parameters")
        return None
    x = np.linspace(0,c,1000)
    y = x**a*np.exp(-b*x)
    return spi.trapz(y,x)
```

```python
exp_integral(10,0,2,1000)
```

```python
exp_integral(1,1,1,-2)
```

## Question 2

```python
def pq_integral(p,q,b):
    if p<0 or q<0 or b<0:
        print("Error: Negative Paramters")
        return None
    def f(x):
        return (x**p)/(1+x**p)**0.5
    I, abserr = spi.quad(f,0,b)
    return I
```

```python
pq_integral(1,2,3)
```

```python
pq_integral(0,0,1)
```

```python
pq_integral(0,-1,1)
```

## Question 3

```python
def sin_cos_fun(a,b,t):
    def f(y,t):
        return np.sin(a*t)+np.cos(b*t**2)
    y = spi.odeint(f,0,t)
    plt.plot(t,y)
```

```python
sin_cos_fun(1,1,np.linspace(0,5,100))
```

```python
sin_cos_fun(-2,0.3,np.linspace(0,20,10000))
```

```python
sin_cos_fun(100,20,np.linspace(0,5,100000))
```

## Question 4

```python
def mathieu(a,q,y0,t):
    def f(u,t):
        return np.array([u[1], -(a-2*q*np.cos(2*t))*u[0]])
    u = spi.odeint(f,y0,t)
    plt.plot(t,u[:,0])
    plt.show()
```

```python
mathieu(7,-2,[0,1],np.linspace(-1,5,1000))
```

## Question 5

```python
def ode_system(a,b,c,u0,t):
    def f(u,t):
        return np.array([a*(u[1]-u[0]), u[0]*(b-u[2]), u[0]*u[1]-c*u[2]])
    u = spi.odeint(f,u0,t)
    plt.plot(u[:,0],u[:,1])
```

```python
ode_system(10,28,3,[1,1,1],np.linspace(0,30,10000))
```

## Question 6

```python
def eigvals_squared(A):
    a_symmetric = A @ A.T
    evals, evecs = la.eig(a_symmetric)
    values = evals.real**2
    return np.sum(values)
```

```python
eigvals_squared(np.array([[5,7],[-6,1]]))
```

```python
eigvals_squared(np.array([[7,4,-3],[7,-6,3],[2,7,-4],[9,5,-2]]))
```

```python
def eigvals_squared2(A):
    M = np.matrix(A)
    eigvals, eigvecs = la.eig(M * M.T)
    return sum(eigvals.real**2)
```

```python
eigvals_squared(np.array([[7,4,-3],[7,-6,3],[2,7,-4],[9,5,-2]]))
```
