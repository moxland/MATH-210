```python
def curve(A,B,C):
    t = np.linspace(0,2*np.pi,1000)
    x = A*np.sin(t)**3
    y = B*np.cos(t) - C*np.cos(2*t)
    plt.plot(x,y,'pink'), plt.axis('equal')
```

## Question 2

```python
def power_series(a,x):
    """Returns the partial power series sum"""
    terms = [a[k] * (x**k) for k in range(0,len(a))]
    return np.array(terms).sum()
```

```python
power_series(np.ones(100),0.5)
```

```python
from scipy.special import factorial
```

```python
power_series(1 / factorial(np.arange(0,100)),1)
```

f_{triangle}(t) = \frac{8}{\pi^2} \sum_{k=0}^{\infty} 
(-1)^k \frac{\sin(2\pi(2k+1)t)}{(2k+1)^2}
$$

b)

```python
def triangle_wave(N,interval):
    """Plots the Nth partial sum of the Fourier series."""
    a = interval[0]
    b = interval[1]
    t = np.linspace(a,b,1000)
    f = (8/(np.pi**2)) * sum([((-1)**k) * np.sin(2*np.pi*(2*k+1)*t)/((2*k+1)**2) for k in range(0,N+1)])
    plt.plot(t,f), plt.grid('on'), plt.ylim(-2,2)
```

```python
triangle_wave(3,[0,3])
```

## Question 4

a)

$$
f_{sawtooth}(t) = \frac{2}{\pi} \sum_{k=1}^{\infty}
\frac{(-1)^{k+1}\sin(2 \pi k t)}{k}
$$

b)

```python
def sawtooth_waves(n,m,T):
    plt.figure(figsize=(2*m,2*n))
    for N in range(1,(n*m)+1):
        plt.subplot(n,m,N)
        t = np.linspace(0,T,100)
        f = (2/np.pi) * sum([(((-1)**(k+1)) * np.sin(2*np.pi*k*t))/k for k in range(1,N+1)])
        plt.plot(t,f), plt.grid('on'), plt.xlim(0,T), plt.title('N= '+str(N)),
        plt.tight_layout()
```

```python
sawtooth_waves(2,4,2)
```

$$
\zeta(s) = \sum_{n=1}^{\infty} \frac{1}{n^s} = \prod_{p \hspace{1mm} prime} \frac{1}{1-p^{-s}}
= \frac{1}{1-2^{-s}} \cdot \frac{1}{1-3^{-s}} \cdot \frac{1}{1-5^{-s}} \cdot 
\frac{1}{1-7^{-s}} \dotsb \frac{1}{1-p^{-s}} \dotsb
$$

b)

```python
def is_prime(N):
    "Determine whether or not N is a prime number."
    if N <= 1:
        return False
    # N is prime of N is only divisble by 1 and itself
    # We should test whether N is divisible by d for all 1 < d < N
    for d in range(2,N):
        # Check if N is divisible by d
        if N % d == 0:
            return False
        # If we exit the for loop, then N is not divisible by any d and N is prime
    return True
```

```python
def euler_product(s,N):
    initial = 1
    for p in range(0,N+1):
        if is_prime(p):
            initial = initial * (1/(1-(p**-s)))
    return initial   
```

```python
euler_product(4,10000)
```

```python
def slope_field(f,tlims,ylims,grid_step):
    y = np.arange(ylims[0],ylims[-1],grid_step)
    t = np.arange(tlims[0],tlims[-1],grid_step)
    L = grid_step/3
    for T in t:
        for Y in y:
            m = f(T,Y)
            theta = np.arctan(m)
            domain = np.linspace((T-L*np.cos(theta)),(T+L*np.cos(theta)),2)
            z = m*(domain-T)+Y
            plt.plot(domain,z,'b'), plt.xlim(tlims[0],tlims[-1]), plt.ylim(ylims[0],ylims[-1])
```

```python
def f(t,y):
    return y**2-t**2
```

```python
slope_field(f,[-3,3],[-2,2],0.2)
```
